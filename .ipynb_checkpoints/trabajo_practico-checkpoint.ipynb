```python
cam = CameraReader()
vid_height, vid_width = cam.get_resolution()
print("Resoluci√≥n del video: {0}x{1}p".format(vid_width, vid_height))
print("Cuadros por segundo: {0}".format(cam.get_fps()))

out = widgets.Image(layout=widgets.Layout(width=str(vid_width), 
                                            height=str(vid_height)))
display(out)
#Mostrar en cuadros
fig, ax = plt.subplots(1, 2, figsize=(16, 9), tight_layout=True)
figb, bx = plt.subplots(1, 2, figsize=(16, 9), tight_layout=True)

#Ajuste del filtro
fx, fy = fftpack.fftfreq(n=1000, d=1), fftpack.fftfreq(n=600, d=1)
Fx, Fy = np.meshgrid(fx, fy)
inv_psf =  np.exp(-2*np.pi**2*(Fx**2 + Fy**2)*2**2)
count = 0;
#Ajuste convolucion
x = np.linspace(-5, 5, num=11)
X, Y = np.meshgrid(x, x)
sigma = 2
psf = np.exp(-0.5*(X)**2/sigma**2 - 0.5*Y**2/sigma**2)/(2.0*np.pi*sigma**2)

#hacer mascara
sigma=200
cy, cx = vid_height/2, vid_width/2
x = np.arange(0, vid_width); y = np.arange(0, vid_height);
X, Y = np.meshgrid(x, y)
mask = 1e-8 + np.exp(-(((X-cx)/sigma)**2 + ((Y-cy)/sigma)**2))

#play = widgets.Play


for frame in cam:
    #print(frame[0])
    # Implemente una funci√≥n que limpie los frames
    #frame_limpio = np.ndarray(frame[0])
    frame = fftconvolve(frame, psf, mode='same') \
    + 20*np.random.randn(frame.shape[0], frame.shape[1])
    frame_sp = (frame.astype('float32'))
    frame_limpio = frame_sp * mask
    #print(sys.getsizeof(frame_limpio))  #getsizeof retorna el tama√±o en Bytes
    
    # Implemente una funci√≥n que comprima los frames
    # frame_comprimido = comprimir(frame)
    # print(sys.getsizeof(frame_comprimido))
    
    # Implemente una funci√≥n que descomprima los frames
    # frame = descomprimir(frame_compromido)
    # print(cam.error(frame))  # Error de compresi√≥n
    count +=1;
    #print(frame)
    #ax[0].imshow(frame); ax[0].set_title("Imagen")
    
    ax[0].imshow(frame);ax[0].set_title("Imagen")
    ax[1].imshow(frame_sp);ax[1].set_title("Espectro")
    
    bx[0].imshow(frame_limpio);bx[0].set_title("Espectro con filtro")
    bx[1].imshow(np.real(frame_limpio));bx[1].set_title("Recuperado")
    
    #display_frame(out, frame)
    #display_frame(out,frame_limpio)
    
    if(count==1):
        break;
    
        
count 
```
